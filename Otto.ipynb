```
import pandas as pd
import numpy as np
import tensorflow as tf
tf.random.set_seed(2019)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
import os
```

```
os.chdir("/content/drive/My Drive/Otto")
```

```
train = pd.read_csv("train.csv",index_col=0)
test = pd.read_csv("test.csv",index_col=0)
sample_sub = pd.read_csv("sampleSubmission.csv",index_col=0)
```

```
train.head()
```

```
test.head()
```

```
print(train["target"].unique())
```

```
dum_train = pd.get_dummies(train)
dum_train.head()
```

```
X = dum_train.iloc[:,:-9]
X.head()
```

```
y = dum_train.iloc[:,-9:]
y.head()
```

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X) 
```

```
(X.shape, y.shape)
```

```
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from keras import metrics
```

```
tf.random.set_seed(200)
glorot_init = tf.keras.initializers.GlorotNormal(seed=200)
```

```
model = tf.keras.models.Sequential([
tf.keras.layers.Dense(97, kernel_initializer=glorot_init, activation='selu'),
tf.keras.layers.BatchNormalization(),
tf.keras.layers.Dense(85, kernel_initializer=glorot_init, activation='selu'),
tf.keras.layers.BatchNormalization(),
tf.keras.layers.Dense(75, kernel_initializer=glorot_init, activation='selu'),
tf.keras.layers.BatchNormalization(),
tf.keras.layers.Dense(65, kernel_initializer=glorot_init, activation='selu'),
tf.keras.layers.BatchNormalization(),
tf.keras.layers.Dense(55, kernel_initializer=glorot_init, activation='selu'),
tf.keras.layers.BatchNormalization(),
tf.keras.layers.Dense(45, kernel_initializer=glorot_init, activation='selu'),
tf.keras.layers.BatchNormalization(),
tf.keras.layers.Dense(45, kernel_initializer=glorot_init, activation='selu'),
tf.keras.layers.BatchNormalization(),
tf.keras.layers.Dense(35, kernel_initializer=glorot_init, activation='selu'),
tf.keras.layers.BatchNormalization(),
tf.keras.layers.Dense(25, kernel_initializer=glorot_init, activation='selu'),
tf.keras.layers.BatchNormalization(),
tf.keras.layers.Dense(15, kernel_initializer=glorot_init, activation='selu'),
tf.keras.layers.BatchNormalization(),
tf.keras.layers.Dense(15, kernel_initializer=glorot_init, activation='selu'),
tf.keras.layers.BatchNormalization(),
tf.keras.layers.Dense(9,activation='softmax')
])
```

```
model.compile(optimizer='adam', loss='categorical_crossentropy',metrics=['accuracy'])
```

```
model.fit( X,y,verbose=2,epochs=200)
```

```
loss,acc =model.evaluate(X,y,verbose=0)
print('Test loss = {:.4f} '.format(loss))
print('Test acc = {:.4f} '.format(acc))
```
